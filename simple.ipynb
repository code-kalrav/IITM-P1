```python
import json
import requests
import pandas as pd
import pickle
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# Base URL
base_url = "https://api.github.com/search/users?q=location:Berlin+followers:%3E200&per_page=100&page="

# List to store all user data
all_users = []

# Loop through each page (saw on the page that there are 600 entries)
for page in range(1, 7):  # Adjust this range if there are more pages
    url = f"{base_url}{page}"
    response = requests.get(url)
    data = response.json()
    users = data.get("items", [])
    all_users.extend(users)  # Add users from this page to the list


# Save all users data to a single JSON file
with open("berlin_users_all_pages.json", "w") as file:
    json.dump(all_users, file, indent=4)

# Print total count of users fetched
print("Total users fetched:", len(all_users))
```

```python
# Define the function to scrape repositories and append to a JSON file
def repo_scrap(url1, headers=None):
    # Initialize page number for pagination
    page1 = 1

    # Loop through paginated results
    while True:
        # Fetch data for the current page
        response = requests.get(f"{url1}?per_page=100&page={page1}", headers=headers)

        # Check if the request was successful
        if response.status_code != 200:
            print(f"Failed to fetch page {page1}, status code: {response.status_code}")
            break

        # Parse JSON data
        repos = response.json()

        # Break the loop if there are no more repositories
        if not repos:
            break

        # Append each repo to the JSON file
        with open('all_repos.json', 'a') as json_file:
            for repo in repos:
                # Append each repository's data to the JSON file on a new line
                json.dump(repo, json_file)
                json_file.write('\n')  # Write each repo as a separate line for easier reading

        # Increment page number
        page1 += 1
```

```python
var = 1
for users in all_users:
    repo_scrap(users["repos_url"], headers)
    print(f"Done For user {var} : {users['login']}") # to check the progress
    var += 1
```
